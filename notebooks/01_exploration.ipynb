# Customer Happiness — Exploration

Walk-through of the dataset and modelling choices. The full reproducible pipeline lives in `src/train.py`.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style='whitegrid')
df = pd.read_csv('../data/ACME-HappinessSurvey2020.csv')
df.head()
```

```python
df.describe().round(2)
```

## Class balance

Slightly more happy than unhappy customers, but close to 50/50 — no need to handle severe imbalance.

```python
df['Y'].value_counts(normalize=True).round(3)
```

## Correlation

Look at which features correlate with the target and with each other.

```python
fig, ax = plt.subplots(figsize=(7, 5))
sns.heatmap(df.corr(), annot=True, fmt='.2f', cmap='RdBu_r', center=0, ax=ax)
ax.set_title('Feature correlations')
plt.show()
```
